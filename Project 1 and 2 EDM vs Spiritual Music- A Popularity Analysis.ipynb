```python
import pandas as pd
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.columns
```

```python
# Look only at tracks that mention "edm" or "spiritual" in their genre
edm_df = df[df["track_genre"].str.contains("edm|spiritual", case=False, na=False)]

# See the first 5 rows
edm_df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(edm_df["danceability"], edm_df["energy"], alpha=0.3, color="purple")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.title("EDM & Spiritual Tracks: Danceability vs Energy")
plt.grid(True)
plt.show()
```

```python
top_tracks = edm_df.sort_values(by="popularity", ascending=False).head(10)
top_tracks[["track_name", "artists", "popularity"]]
```

```python
edm_df.groupby("track_genre")["popularity"].mean().sort_values(ascending=False)
```

```python
# Group by genre and calculate average popularity
genre_popularity = edm_df.groupby("track_genre")["popularity"].mean().sort_values(ascending=False)

# Peek at it
genre_popularity.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
genre_popularity.plot(kind='bar', color='orange', edgecolor='black')

plt.title("Average Popularity by Genre (EDM & Spiritual Tracks)")
plt.xlabel("track_Genre")
plt.ylabel("Average Popularity")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
bars = plt.bar(genre_popularity.index, genre_popularity.values, color='skyblue', edgecolor='black')

plt.title("Average Popularity by Genre (EDM & Spiritual Tracks)")
plt.xlabel("track_Genre")
plt.ylabel("Average Popularity")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text labels above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Focus on genres you're interested in
target_genres = ["edm", "spiritual", "house", "trance", "meditation"]
mask = df["track_genre"].str.contains("|".join(target_genres), case=False, na=False)
music_df = df[mask].copy()

# Preview it
music_df.head()
```

```python
# Group by genre and calculate averages
genre_features = music_df.groupby("track_genre")[["danceability", "energy", "valence", "tempo"]].mean().sort_values("danceability", ascending=False)

# View the summary
genre_features
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Plot all three features
genre_features[["danceability", "energy", "valence"]].plot(kind="bar", figsize=(12,6), colormap="plasma")

plt.title("ðŸ’ƒ Musical Characteristics by Genre (EDM & Spiritual)")
plt.xlabel("Genre")
plt.ylabel("Average Value")
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
for genre in music_df["track_genre"].unique():
    genre_data = music_df[music_df["track_genre"] == genre]
    plt.scatter(genre_data["energy"], genre_data["valence"], label=genre, alpha=0.4)

plt.title("ðŸ§  Energy vs Valence â€” How Music Feels")
plt.xlabel("Energy")
plt.ylabel("Valence (Happiness)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
